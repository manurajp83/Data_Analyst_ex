```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv('iris.csv')
data.head()
```

```python
sns.lineplot(x='SepalLengthCm',y='SepalWidthCm',data=data)
plt.title('Sepal length v/s Sepal width')
```

##### Styling plots

##### set_style()
5 themes are available :white, dark, whitegrid, darkgrid, ticks

```python
sns.set_style('white')
sns.lineplot(x='SepalLengthCm',y='SepalWidthCm',data=data)
plt.title('Sepal length v/s Sepal width')
```

```python
sns.set_style('dark')
sns.lineplot(x='SepalLengthCm',y='SepalWidthCm',data=data)
plt.title('Sepal length v/s Sepal width')
```

```python
sns.set_style('ticks')
sns.lineplot(x='SepalLengthCm',y='SepalWidthCm',data=data)
plt.title('Sepal length v/s Sepal width')
```

```python
sns.set_style('whitegrid')
sns.lineplot(x='SepalLengthCm',y='SepalWidthCm',data=data)
plt.title('Sepal length v/s Sepal width')
```

```python
sns.set_style('darkgrid')
sns.lineplot(x='SepalLengthCm',y='SepalWidthCm',data=data)
plt.title('Sepal length v/s Sepal width')
```

##### despine 

Remove borders.Remove top and right borders by default

```python
sns.set_style('darkgrid')
sns.lineplot(x='SepalLengthCm',y='SepalWidthCm',data=data)
plt.title('Sepal length v/s Sepal width')
sns.despine()
```

```python
# Size

plt.figure(figsize=(3,3))
sns.lineplot(x='SepalLengthCm',y='SepalWidthCm',data=data)
plt.title('Sepal length v/s Sepal width')
sns.despine(left=True)
```
