```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('iris.csv')
```

```python
sns.set_style('dark')
sns.scatterplot(x='SepalLengthCm',y='SepalWidthCm',data=data)
```

```python
sns.set_style('dark')
sns.scatterplot(x='SepalLengthCm',y='SepalWidthCm',data=data,hue='SepalWidthCm')

```

```python
sns.set_style('dark')
sns.scatterplot(x='SepalLengthCm',y='SepalWidthCm',data=data,hue='SepalWidthCm',palette='gist_earth')

```

```python
sns.set_style('dark')
sns.scatterplot(x='SepalLengthCm',y='SepalWidthCm',data=data,hue='SepalWidthCm',palette='gist_earth',legend='full')
```

```python
sns.set_style('dark')
sns.scatterplot(x='SepalLengthCm',y='SepalWidthCm',data=data,hue='SepalWidthCm',palette='gist_earth',legend=False)
```

```python
sns.set_style('dark')
sns.scatterplot(x='SepalLengthCm',y='SepalWidthCm',data=data,hue='SepalWidthCm',palette='gist_earth',alpha=0.5)
# alpha defines the opacity of the points
```

```python


#### bar plot
titanic=pd.read_csv(r'titanic.csv')
titanic
```

```python
# the height of each rectangle will be the measure of central tendency of each x variable

sns.barplot(x='Pclass',y='Fare',data=titanic)
```

```python
sns.barplot(x='Pclass',y='Fare',hue='Sex',data=titanic)
```

```python
#HORIZONTAL BARPLOT
sns.barplot(x='Fare',y='Sex',data=titanic)
```

```python
# PLOT BARS IN A GIVEN ORDER
sns.barplot(x='Pclass',y='Fare',order =[3,1,2],data=titanic)
```

```python
# PLOT BARS with same color
sns.barplot(x='Pclass',y='Fare',order =[3,1,2],data=titanic,color='salmon')
```

## countplot

```python
sns.countplot(x='Sex',data=titanic)
```

```python
sns.countplot(x='Sex',hue='Pclass', data=titanic)
#hue used to split the plot
```

```python
sns.countplot(x='Sex',hue='Pclass', data=titanic,palette='Purples')
```

```python
sns.pairplot(titanic,hue='Sex')
## univariate same variable (a,a) (b,b)
```
