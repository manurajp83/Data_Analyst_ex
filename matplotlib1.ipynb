# Pie chart

```python
import matplotlib.pyplot as plt
slices=[30,20,40]
label=['women','men','kids']
plt.style.use('classic')
plt.pie(slices,labels=label)
plt.title('PIE CHART')
```

```python
# custom colours and properties
slices = [30,40,20]
labels = ['Men','Women','Kids']
col=['cyan','purple','blue']
plt.pie(slices,labels=labels,colors=col,wedgeprops={'linewidth':2,'edgecolor':'white'})
plt.title('PIE CHART')
```

```python
# Explodes and shadows
slices = [30,40,20]
labels = ['Men','Women','Kids']
col=['cyan','purple','blue']
explodes=[0,0,.1]
plt.pie(slices,labels=labels,colors=col,wedgeprops={'linewidth':2,'edgecolor':'white'},explode=explodes, shadow=True)
plt.title('PIE CHART')
```

```python
# start angle and percentage value
slices = [30,40,20]
labels = ['Men','Women','Kids']
col=['cyan','purple','blue']
explodes=[0,0,.1]
plt.pie(slices,labels=labels,colors=col,wedgeprops={'linewidth':2,'edgecolor':'white'},explode=explodes, shadow=True,startangle=90,autopct='%.2f%%')

plt.title('PIE CHART')
```

```python
import numpy as np
a= np.random.randint(1,10,size=9)
b= np.random.randint(2,15,size=9)
plt.style.use('seaborn')
print(a)
print(b)
plt.scatter(a,b)
```

```python
#scatter plot
import numpy as np
a = np.random.randint(1,10, size=9)
b = np.random.randint(2,15, size=9)
plt.style.use('seaborn')
plt.scatter(a,b, s=150 ,color='pink',marker='o',edgecolors='black',linewidth=2)
```

```python
a = np.random.randint(1,10, size=9)
b = np.random.randint(2,15, size=9)
plt.style.use('seaborn')
print(a)
print(b)
colors=b
plt.scatter(a,b, s=150 ,marker='s',c=colors,cmap='Purples',edgecolors='black',linewidth=2)
cbar=plt.colorbar()
cbar.set_label('color variation')
```

```python
# stack plot
days =[1,2,3,4,5]
studying=[6,8,7,6,11]
playing=[5,8,7,6,13]
plt.stackplot(days,studying,playing,colors=['red','pink'])
plt.xlabel('Days')
plt.ylabel('Number of hours')
plt.legend(['studying','Playing'])
```
