## Groupby
creates sub dataframes based on a single column value or a combination of multiple column values

```python
import pandas as pd
df=pd.DataFrame({'Name':['A','B','C','D','E'],
                'Sex':['M','F','F','M','F'],
                'Age':[23,21,34,38,32]})
df
```

```python
grp=df.groupby('Sex')
grp
```

```python
grp.apply(display)
```

```python
data=pd.read_csv("weather_data.csv")
data
```

```python
grp=data.groupby("city")
grp
```

```python
grp.apply(display)
```

```python
for city,city_df in grp:
    print(city)
    print(city_df)
```

```python
grp.get_group('paris')
```

```python
grp.min()
```

```python
grp.max()
```

```python
grp.get_group("paris").max()
```

```python
#groupby using multiple columns
grp=data.groupby(['city','event'])
grp.apply(display)
```

```python
grp.get_group(('newyork','Sunny'))
```

## concat()

```python
tcs={'id':[101,102,103,104],'Name':['Max','Dipin','Jose','Rahul'],'Age':[25,26,27,29]}
wipro={'id':[101,102,103,104],'Name':['Dev','Sreenath','Sreerag','shafeel'],'Age':[24,24,27,26]}
```

```python
tcs_emp=pd.DataFrame(tcs)
wipro_emp=pd.DataFrame(wipro)
tcs_emp
```

```python
pd.concat([tcs_emp,wipro_emp])
```

```python
pd.concat([tcs_emp,wipro_emp],ignore_index=True)
```

```python
pd.concat([tcs_emp,wipro_emp],keys=['TCS','WIPRO'])
```

```python
# concat horizontally
pd.concat([tcs_emp,wipro_emp], axis = 1)
```

```python
#concatenating a datatframe with a series
salary=pd.Series([20000,23000,24000,21000],name='salary')
pd.concat([tcs_emp,salary],axis=1)
```

```python
tcs={'id':[101,102,103,104],'Name':['Max','Dipin','Jose','Rahul'],'Age':[25,26,27,29]}
wipro={'id':[101,102,103,106],'Name':['Dev','Sreenath','Sreerag','shafeel'],'salary':[25000,34000,37000,20000]}

tcs_emp=pd.DataFrame(tcs)
wipro_emp = pd.DataFrame(wipro)
```

```python
tcs_emp
```

```python
wipro_emp
```

```python
pd.merge(tcs_emp,wipro_emp, on='id') #inner merging -- selects key common to both the dataframe
```

```python
pd.merge(tcs_emp,wipro_emp,on='id', suffixes=['_left','_right'])
```

```python
pd.merge(tcs_emp,wipro_emp,on='id',how='outer') #outer means union
```

```python
pd.merge(tcs_emp,wipro_emp,on='id',how='outer',indicator=True)
```

```python
pd.merge(tcs_emp,wipro_emp,on='id',how='left',indicator=True)
```

## Map

```python
def square(x):
    return x**2

my_list=[6,5,4,3]
list(map(square,my_list))
```

```python
data=pd.read_csv('weather_data.csv')
data
```

```python
def square(x):
    return x**2
data['temperature'].map(square)
```

```python
data_copy=data.copy()
data_copy['new_temperature']=data['temperature'].map(square)
data_copy
```

```python
#for handling multiple columns simultaneously
data[['temperature','windspeed']].applymap(square)
```
