## Handling missing vaues

```python
data_missing=pd.read_csv('weather_data_missing.csv')
data_missing
```

```python
#check for missing values
data_missing.isna()
```

```python
data_missing.isna().sum()
```

```python
data_missing.isnull().sum()
```

```python
#fill missing values
data2 = data_missing.fillna(0)
data2
```

```python
avg=data_missing['temperature'].mean()
data_missing.fillna(avg)
```

```python
data_missing.fillna(method='ffill')# forward fill
```

```python
data_missing.fillna(method='bfill')# backward fill
```

```python
#linear inetrpolate--- fill ith average of preceeding and suceeding values
data_missing.interpolate()
```

```python
#drop rows with null values
data_missing.dropna()
```

```python
# drop rows with full NaN values
data_missing.dropna(how='all')
```
