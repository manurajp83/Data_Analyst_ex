## Convereters

```python
#converter method
def convert_people(name):
    if name=='larry page':
        return 'Ben'
    else:
        return name
    
data_convert= pd.read_excel('stock_data.xlsx',converters={'people':convert_people})   
data_convert
```

```python
data_convert.to_excel("convert_people.xlsx")
```
